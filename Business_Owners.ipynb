```python
import pandas as pd
from sqlalchemy import create_engine
pd.set_option('display.max_columns', None)
```

```python
csv_file = "Business_Owners.csv"
owners_df = pd.read_csv(csv_file)
owners_df.head()
```

```python
csv_file = "Business_Licenses_-_Current_Active.csv"
licenses_df = pd.read_csv(csv_file)
licenses_df.head()
```

```python
licenses_df = licenses_df.rename(columns={"ACCOUNT NUMBER": "Account Number"})
licenses_df.head()
```

```python

overall = pd.merge(licenses_df, owners_df, how="left", on="Account Number")
overall.head()
```

```python
list(overall.columns)
```

```python
overall.columns = overall.columns.str.upper()
```

```python
overall = overall.drop(columns=['LEGAL NAME',
                      'SITE NUMBER',
                      'WARD',
                      'WARD PRECINCT',
                      'POLICE DISTRICT',
                      'LICENSE CODE',
                      'BUSINESS ACTIVITY ID',
                      'LICENSE STATUS CHANGE DATE',
                      'SSA',
                      'LATITUDE',
                      'LONGITUDE',
                      'LOCATION',
                      'LEGAL NAME',
                      'OWNER MIDDLE INITIAL',
                      'SUFFIX',
                      'LEGAL ENTITY OWNER'])
```

```python
list(overall.columns)
```

```python
overall = overall.rename(columns={'ID': "ID",
                                  'LICENSE ID': 'LICENSE_ID',
                                     'ACCOUNT NUMBER': 'ACCOUNT_NUMBER',
                                     'DOING BUSINESS AS NAME': 'DOING_BUSINESS_AS_NAME',
                                     'ZIP CODE':'ZIP_CODE',
                                     'LICENSE DESCRIPTION':'LICENSE_DESCRIPTION',
                                     'BUSINESS ACTIVITY':'BUSINESS_ACTIVITY',
                                     'LICENSE NUMBER':'LICENSE_NUMBER',
                                     'APPLICATION TYPE':'APPLICATION_TYPE',
                                     'APPLICATION CREATED DATE':'APPLICATION_CREATED_DATE',
                                     'APPLICATION REQUIREMENTS COMPLETE':'APPLICATION_REQUIREMENTS_COMPLETE',
                                     'PAYMENT DATE':'PAYMENT_DATE',
                                     'CONDITIONAL APPROVAL':'CONDITIONAL_APPROVAL',
                                     'LICENSE TERM START DATE':'LICENSE_TERM_START_DATE',
                                     'LICENSE TERM EXPIRATION DATE':'LICENSE_TERM_EXPIRATION_DATE',
                                     'LICENSE APPROVED FOR ISSUANCE':'LICENSE_APPROVED_FOR_ISSUANCE',
                                     'DATE ISSUED':'DATE_ISSUED',
                                     'LICENSE STATUS':'LICENSE_STATUS',
                                     'OWNER FIRST NAME':'OWNER_FIRST_NAME',
                                     'OWNER LAST NAME':'OWNER_LAST_NAME'})
```

```python
overall.columns = overall.columns.str.lower()
overall.head()
```

```python
overall.dtypes
```

```python
overall.set_index("id", inplace=True)
```

```python
rds_connection_string = "postgres:Falcons01!@localhost:5432/business_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
overall.to_sql(name='overall', con=engine, if_exists='append', index=True)
```
